```python
import numpy as np
import matplotlib.pyplot as plt

# Load data while skipping comment lines starting with "#"
data = np.loadtxt('SPINTEXTURE_2D_GAMMA_CENTER.dat', comments='#')

# Extract columns
kx = data[:, 0]
ky = data[:, 1]
energy = data[:, 2]
sx = data[:, 3]
sy = data[:, 4]
sz = data[:, 5]

# Infer grid size (e.g., 30x30)
grid_size = int(np.sqrt(len(kx)))  # or manually set to 30 if it's not square

# Reshape data for grid plotting
kx_grid = kx.reshape((grid_size, grid_size))
ky_grid = ky.reshape((grid_size, grid_size))
sx_grid = sx.reshape((grid_size, grid_size))
sy_grid = sy.reshape((grid_size, grid_size))
sz_grid = sz.reshape((grid_size, grid_size))

# Normalize arrows for better visual clarity (optional)
norm = np.sqrt(sx_grid**2 + sy_grid**2)
norm[norm == 0] = 1  # avoid divide-by-zero
sx_norm = sx_grid / norm
sy_norm = sy_grid / norm

# Plot spin texture (Sx, Sy) as arrows
plt.figure(figsize=(8, 6))
plt.quiver(kx_grid, ky_grid, sx_norm, sy_norm, sz_grid, cmap='coolwarm', scale=30)
plt.xlabel('k')
plt.ylabel('k')
plt.title('Spin Texture (S, S) colored by S_z')
plt.colorbar(label='S_z')
plt.xlim(min(kx), max(kx))
plt.ylim(min(ky), max(ky))
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load data
data = np.loadtxt("SPINTEXTURE_2D_GAMMA_CENTER.dat", comments="#")
kx, ky, energy, sx, sy, sz = data.T

# Create 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(projection='3d')
p = ax.scatter(kx, ky, energy, c=sz, cmap='coolwarm', marker='o')

ax.set_xlabel("k")
ax.set_ylabel("k")
ax.set_zlabel("Energy (eV)")
ax.set_title("3D Band Structure with Spin-z Coloring")
fig.colorbar(p, ax=ax, label="S_z")
ax.view_init(elev=90, azim=-90)  # Adjust view angle for better visibility
plt.tight_layout()
plt.show()

```
