```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np


data = np.loadtxt('bands.dat.gnu')

fermi = -0.9123

k = np.unique(data[:, 0])
bands = np.reshape(data[:, 1], (-1, len(k)))

plt.figure(figsize=(14, 8))

for band in range(len(bands)):
    plt.plot(k, bands[band, :]-fermi, linewidth=1, alpha=0.7, color='red')
plt.xlim(min(k), max(k))

plt.axhline(0,lw=0.5,ls='--',color='k')

plt.axvline(0.6724, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(1.2550, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(1.5907, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(2.2631, linewidth=0.75, color='k', alpha=0.5)

plt.xticks(ticks= [0, 0.6724, 1.2550, 1.5907, 2.2631], \
           labels=['K','$\Gamma$', 'M', 'K', '$\Gamma$'])
plt.ylabel("Energy (eV)")
plt.ylim(-5,5)
plt.show()
```
