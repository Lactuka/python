```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data_bi = pd.read_csv(r"PBAND_A1_SOC.dat", comment='#', delim_whitespace=True, header=None, skiprows=2)
data_sb = pd.read_csv(r"PBAND_A2_SOC.dat", comment='#', delim_whitespace=True, header=None, skiprows=2)

k = np.array(data_bi[0])
e = np.array(data_bi[1])
s = np.array([[value, 'blue'] for value in data_bi[2]], dtype=object)
s = np.transpose(s)
px = np.array([[value, 'red'] for value in data_sb[3]], dtype=object)
px = np.transpose(px)
py = np.array([[value, 'green'] for value in data_sb[4]], dtype=object)
py = np.transpose(py)
pz = np.array([[value, 'orange'] for value in data_sb[5]], dtype=object)
pz = np.transpose(pz)

plt.figure(figsize=(14, 8))

# Track which labels we've already added
labels_added = set()

for i in range(len(s[0])):
    # Create pairs of (value, color) for this index
    sorting = [
        (k[i], e[i], s[0][i], s[1][i], 's'),      # Added orbital name
        (k[i], e[i], px[0][i], px[1][i], 'px'),    
        (k[i], e[i], py[0][i], py[1][i], 'py'),    
        (k[i], e[i], pz[0][i], pz[1][i], 'pz')     
    ]
    
    data = sorted(sorting, key=lambda x: x[2], reverse=True)
    
    # Only add label if we haven't seen this orbital type before
    orbital_name = data[0][4]
    if orbital_name not in labels_added:
        plt.scatter(data[0][0], data[0][1], color=data[0][3], s=10, label=orbital_name)
        labels_added.add(orbital_name)
    else:
        plt.scatter(data[0][0], data[0][1], color=data[0][3], s=10)

plt.ylim(-5,5)
plt.xlim(min(k), max(k)) 
plt.ylabel(r'$E-E_f$', fontsize=16)
plt.legend(frameon=True, loc='upper right', fontsize=12)
plt.axhline(0, ls='--', lw=0.5, color='black')
plt.axvline(0.972, ls='--', lw=0.5, color='black')
plt.axvline(1.814, ls='--', lw=0.5, color='black')
plt.axvline(2.300, ls='--', lw=0.5, color='black')

plt.xticks(ticks=[0, 0.972, 1.814, 2.300, 3.273], labels=['K', '$\Gamma$', 'M', 'K', '$\Gamma$'])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data_bi = pd.read_csv(r"PBAND_A1_SOC.dat", comment='#', delim_whitespace=True, header=None, skiprows=2)
data_sb = pd.read_csv(r"PBAND_A2_SOC.dat", comment='#', delim_whitespace=True, header=None, skiprows=2)

k = np.array(data_bi[0])
e = np.array(data_bi[1])
s = np.array([[value, 'gray'] for value in data_bi[2]], dtype=object)
s = np.transpose(s)
px = np.array([[value, 'blue'] for value in data_bi[3]], dtype=object)
px = np.transpose(px)
py = np.array([[value, 'green'] for value in data_bi[4]], dtype=object)
py = np.transpose(py)
pz = np.array([[value, 'red'] for value in data_bi[5]], dtype=object)
pz = np.transpose(pz)

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12), sharex=True, sharey=True)
axes = axes.flatten()

orbitals = [s, px, py, pz]
# Different sequential colormaps for each orbital
colormaps = ['Greys', 'Blues', 'Greens', 'Reds']
names = ['s', 'px', 'py', 'pz']

for idx, (orbital, cmap, name) in enumerate(zip(orbitals, colormaps, names)):
    ax = axes[idx]
    
    # Use orbital contributions with sequential colormap
    # Higher contribution = more intense color
    scatter = ax.scatter(k, e, c=orbital[0], s=15, cmap=cmap, alpha=0.8, vmin=0)
    
    ax.set_title(f'{name} orbital contribution', fontsize=14, fontweight='bold')
    ax.set_ylim(-5, 5)
    ax.set_xlim(min(k), max(k))
    
    # Add colorbar for each subplot
    cbar = plt.colorbar(scatter, ax=ax, label='Contribution')
    cbar.ax.tick_params(labelsize=10)
    
    # Add the same grid lines to each subplot
    ax.axhline(0, ls='--', lw=0.5, color='black')
    ax.axvline(0.972, ls='--', lw=0.5, color='black')
    ax.axvline(1.814, ls='--', lw=0.5, color='black')
    ax.axvline(2.300, ls='--', lw=0.5, color='black')
    
    # Add x-axis labels only to bottom subplots
    if idx >= 2:
        ax.set_xticks([0, 0.972, 1.814, 2.300, 3.273])
        ax.set_xticklabels(['K', '$\Gamma$', 'M', 'K', '$\Gamma$'])
    
    # Add y-axis label to left subplots
    if idx % 2 == 0:
        ax.set_ylabel(r'$E-E_f$', fontsize=16)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data_bi = pd.read_csv(r"PBAND_A1_SOC.dat", comment='#', delim_whitespace=True, header=None, skiprows=2)
data_sb = pd.read_csv(r"PBAND_A2_SOC.dat", comment='#', delim_whitespace=True, header=None, skiprows=2)

k = np.array(data_bi[0])
e = np.array(data_bi[1])
s = np.array([[value, 'grey'] for value in data_bi[2]], dtype=object)
s = np.transpose(s)
px = np.array([[value, 'blue'] for value in data_bi[3]], dtype=object)
px = np.transpose(px)
py = np.array([[value, 'green'] for value in data_bi[4]], dtype=object)
py = np.transpose(py)
pz = np.array([[value, 'red'] for value in data_bi[5]], dtype=object)
pz = np.transpose(pz)

plt.figure(figsize=(14, 8))

orbitals = [s, px, py, pz]
colormaps = ['Greys', 'Blues', 'Greens', 'Reds']
names = ['s', 'px', 'py', 'pz']

# Plot each orbital with its own colormap
for orbital, cmap, name in zip(orbitals, colormaps, names):
    # Only plot points where contribution is significant to avoid clutter
    mask = orbital[0] > 0.05  # Adjust threshold as needed
    
    if np.any(mask):  # Only plot if there are significant contributions
        scatter = plt.scatter(k[mask], e[mask], c=orbital[0][mask], 
                            s=20, cmap=cmap, alpha=0.7, 
                            label=f'{name} orbital', vmin=0)

plt.ylim(-5, 5)
plt.xlim(min(k), max(k))
plt.ylabel(r'$E-E_f$', fontsize=16)

# Add grid lines
plt.axhline(0, ls='--', lw=0.5, color='black')
plt.axvline(0.972, ls='--', lw=0.5, color='black')
plt.axvline(1.814, ls='--', lw=0.5, color='black')
plt.axvline(2.300, ls='--', lw=0.5, color='black')

plt.xticks(ticks=[0, 0.972, 1.814, 2.300, 3.273], 
           labels=['K', '$\Gamma$', 'M', 'K', '$\Gamma$'])

# Create a custom legend showing the color schemes
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='grey', markersize=8, label='s orbital'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=8, label='px orbital'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=8, label='py orbital'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=8, label='pz orbital')
]
plt.legend(handles=legend_elements, frameon=True, loc='upper right', fontsize=12)
plt.title('Orbital Contributions of Bi in Band Structure', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
