```python
import numpy as np
import matplotlib.pyplot as plt

# Load data while skipping comment lines starting with "#"
data = np.loadtxt('SPINTEXTURE_2D_GAMMA_CENTER.dat', comments='#')

# Extract columns
kx = data[:, 0]
ky = data[:, 1]
energy = data[:, 2]
sx = data[:, 3]
sy = data[:, 4]
sz = data[:, 5]

# Infer grid size (e.g., 30x30)
grid_size = int(np.sqrt(len(kx)))  # or manually set to 30 if it's not square

# Reshape data for grid plotting
kx_grid = kx.reshape((grid_size, grid_size))
ky_grid = ky.reshape((grid_size, grid_size))
sx_grid = sx.reshape((grid_size, grid_size))
sy_grid = sy.reshape((grid_size, grid_size))
sz_grid = sz.reshape((grid_size, grid_size))

# Normalize arrows for better visual clarity (optional)
norm = np.sqrt(sx_grid**2 + sy_grid**2)
norm[norm == 0] = 1  # avoid divide-by-zero
sx_norm = sx_grid / norm
sy_norm = sy_grid / norm

# Plot spin texture (Sx, Sy) as arrows
plt.figure(figsize=(8, 6))
plt.quiver(kx_grid, ky_grid, sx_norm, sy_norm, sz_grid, cmap='coolwarm', scale=30)
plt.xlabel('kₓ')
plt.ylabel('kᵧ')
plt.title('Spin Texture (Sₓ, Sᵧ) colored by S_z')
plt.colorbar(label='S_z')
plt.xlim(min(kx), max(kx))
plt.ylim(min(ky), max(ky))
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load data
data = np.loadtxt("SPINTEXTURE_2D_GAMMA_CENTER.dat", comments="#")
kx, ky, energy, sx, sy, sz = data.T

# Create 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(projection='3d')
p = ax.scatter(kx, ky, energy, c=sz, cmap='coolwarm', marker='o')

ax.set_xlabel("kₓ")
ax.set_ylabel("kᵧ")
ax.set_zlabel("Energy (eV)")
ax.set_title("3D Band Structure with Spin-z Coloring")
fig.colorbar(p, ax=ax, label="S_z")
ax.view_init(elev=90, azim=-90)  # Adjust view angle for better visibility
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

dos = np.loadtxt("PDOS_C.dat", comments="#")
energy, s, py, pz, px, dxy, dyz, dz2, dxz, dx2y2, tot = dos.T 
plt.figure(figsize=(14, 8))
plt.plot(energy, s, label='s', color='blue')
plt.plot(energy, py, label='p_y', color='orange')
plt.plot(energy, pz, label='p_z', color='green')
plt.plot(energy, px, label='p_x', color='red')
plt.plot(energy, dxy, label='d_{xy}', color='purple')
plt.plot(energy, dyz, label='d_{yz}', color='brown')
plt.plot(energy, dz2, label='d_{z^2}', color='pink')
plt.plot(energy, dxz, label='d_{xz}', color='cyan')
plt.plot(energy, dx2y2, label='d_{x^2-y^2}', color='magenta')
plt.legend()
plt.axvline(0, ls='--', lw=0.5, color='black')
plt.xlabel('Energy (eV)')
plt.ylabel('Projected DOS')
plt.title('Projected Density of States (PDOS)')

```
